```python
from gabotrack import YOLOutils
from gabotrack import Analysis
```

```python
def testImageFile(imgInPath):    
    modelos = [
        YOLOutils.__YOLOV11m__,
        #YOLOutils.__CUSTOM_FRUITS__
   ]

    analysis = Analysis(imgInPath, modelNames=modelos, class_name=None)
    analysis.__saveImageDetection__ = True

    analysis.__determineQRPlane__ = True
    analysis.__showQRProyectivePlane__ = True  
    analysis.__showQRBoundingBoxes__ = False
    analysis.__detectQR_conf__ = 0.6
    analysis.__detectPerson_conf__ = 0.1
    analysis.__showPlaneVertices__ = True

    analysis.__cartesian_length__ = 162
    analysis.__cartesian_width__  = 84.5
    analysis.__cartesian_in_centroid_qr__ = False

    analysis.__showBoundingBoxes__ = False
    analysis.__showBoxLabel__ = False
    
    analysis.__showCartesian__ = True
    
    
    # Inicia el análisis de la imagen
    analysis.begin()
    analysis.end()
```

```python
testImageFile('extras/imgs/qr/mesa05.jpg' )
```
