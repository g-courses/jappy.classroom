### Gráfico con Numpy y Matplotlib

Este ejemplo muestra una imagen que tiene más de un diagrama. Aprovecha las particularidades de un cuaderno Jupyter de escribit código en celdas y de permitir incluir celdas en Markdown para insertar comentarios o aclaraciones.

````python
import numpy as np
import matplotlib.pyplot as plt

def grafico():
    # Parámetros
    fs = 1000  # Frecuencia de muestreo
    t = np.linspace(0, 1, fs)  # Tiempo de 0 a 1 segundo

    # Frecuencias
    f_m = 2  # Frecuencia de la señal moduladora (Hz)
    f_c = 10  # Frecuencia de la señal portadora (Hz)

    # Señales
    moduladora = np.sin(2 * np.pi * f_m * t)  # Señal moduladora
    portadora = np.sin(2 * np.pi * f_c * t)  # Señal portadora

    # Señal modulada
    modulada = (1 + moduladora) * portadora  # Modulación de amplitud

    # Gráficos
    plt.figure(figsize=(12, 6))

    # Señal moduladora
    plt.subplot(3, 1, 1)
    plt.plot(t, moduladora, label='Señal Moduladora (f_m=2 Hz)')
    plt.title('Señal Moduladora')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid()
    plt.legend()

    # Señal portadora
    plt.subplot(3, 1, 2)
    plt.plot(t, portadora, label='Señal Portadora (f_c=10 Hz)', color='orange')
    plt.title('Señal Portadora')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid()
    plt.legend()

    # Señal modulada
    plt.subplot(3, 1, 3)
    plt.plot(t, modulada, label='Señal Modulada', color='green')
    plt.title('Señal Modulada')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid()
    plt.legend()

    plt.tight_layout()
    
    plt.show('senal_modulada.png')  # Guarda la imagen en un archivo
    plt.close()  # Cierra la figura
````

Una vez definida la función `grafico()`, se puede llamar desde otra celda.

````python
grafico()
````

## Realizar diagrama UML con `Mermaid`

En una celda Markdown se debe incluir:

><pre>
>```mermaid
>
># Código Mermaid
>
>```
></pre>

Por ejemplo, la siguiente celda tiene el siguiente código:

><pre>
>```mermaid
>classDiagram
>    class Figura{
>        init()
>
>        area(): double
>    }
>
>    class Rectangulo{
>        base: double
>        altura: double
>
>        int(_base: double, _altura: double)
>        area(): double
>    }
>
>    class Elipse
>    class Cuadrado
>    class Circulo
>
>    direction BT
>    Figura <|-- Rectangulo
>    Rectangulo <|-- Cuadrado
>
>    Figura <|-- Elipse
>    Elipse <|-- Circulo
>```
></pre>

La directiva `direction BT` establece la dirección del renderizado del diagrama. Las opciones son:

* BT: Bottom to Top
* TB: Top to Bottom
* LR: Left to Right
* RL: Right to Left



```mermaid

classDiagram

    class Figura{
        init()

        area(): double
    }

    class Rectangulo{
        base: double
        altura: double

        int(_base: double, _altura: double)
        area(): double
    }

    class Elipse
    class Cuadrado
    class Circulo

    direction BT
    Figura <|-- Rectangulo
    Rectangulo <|-- Cuadrado

    Figura <|-- Elipse
    Elipse <|-- Circulo
```

**Observación**: Cabe destacar crear diagramas a través de celdas `Markdown` utilizando `Mermaid`, es independiente del kernel utilizado. Al igual que las imágenes generadas por código en el cuaderno, al exportarlo como documento html quedan incrustadas en el mismo.

## Librerías Python instaladas

````python
import importlib.metadata

# Listar todas las distribuciones instaladas
installed_packages = importlib.metadata.distributions()

# Crear una lista de tuplas (nombre, versión)
packages_list = [(package.metadata['Name'], package.version) for package in installed_packages]

# Ordenar la lista alfabéticamente por el nombre del paquete
sorted_packages = sorted(packages_list, key=lambda x: x[0].lower())

# Imprimir el nombre y la versión de cada paquete
for name, version in sorted_packages:
    print(f"{name}=={version}")
````
