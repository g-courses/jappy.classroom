# Ejemplos de visualización con ggplot2

## Basic Scatter Plot with Random Data

```r
library(ggplot2)

x_values = runif(100, min = 0, max = 10)
y_values = rnorm(100, mean = 5, sd = 2)
random_data = data.frame(x = x_values, y = y_values)

# Crear un gráfico de dispersión (scatter)
ggplot(random_data, aes(x = x, y = y)) +
  geom_point(color = "blue", alpha = 0.7) +
  labs(title = "Gráfico de dispersión",
       x = "X",
       y = "Y") +
  theme_minimal()
```

* `runif(100, min = 0, max = 10)`: generates 100 random numbers uniformly distributed between 0 and 10.
* `rnorm(100, mean = 5, sd = 2)`: generates 100 random numbers from a normal distribution with a mean of 5 and a standard deviation of 2.
* `ggplot()`: initializes the plot, mapping x and y to the aesthetics.
* `geom_point()`: adds the points to the plot.
* `labs()`: adds a title and axis labels.
* `theme_minimal()`: applies a clean, minimalist theme.

## Histograma de datos aleatorios normales

```r
library(ggplot2)

normal_data = data.frame(value = rnorm(5000, mean = 10, sd = 5))

# Crear un histograma con los datos
ggplot(normal_data, aes(x = value)) +
  geom_histogram(binwidth = 0.5, fill = "skyblue", color = "black") +
  labs(title = "Histograma de datos aletorios normales",
       x = "Valores",
       y = "Cantidad") +
  theme_classic()
```

* `rnorm(5000, mean = 10, sd = 5)`:
genera 5000 números aleatorios desde una distribución normal, cuyo promedio es 10 y su desviación estándar es 5.

* `geom_histogram()`: función que crea el histograma.
* `binwidth`: controla el ancho de las barras del histograma.
* `fill`: configura el color de las barras.
* `color`: configura el color del borde de las barras.
